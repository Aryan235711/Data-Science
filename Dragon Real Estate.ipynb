```python
from sklearn.datasets import fetch_openml
import pandas as pd

# Fetch dataset
boston = fetch_openml(name="boston", version=1, as_frame=True)

# Convert to DataFrame
df = boston.frame

# Save to CSV
df.to_csv("boston_housing.csv", index=False)
```
