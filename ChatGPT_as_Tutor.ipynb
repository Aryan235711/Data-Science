```python
cardinality_threshold = 15 # Set the cardinality count so that columns with massive unique values will be rejected
def isCardinallyLimited(dataset,category, cardinality_threshold):
    categorical_columns = []
    for column in category:
        if dataset[column].nunique() <= cardinality_threshold:
            categorical_columns.append(column)
    return categorical_columns
categorical_columns = isCardinallyLimited(df_sales,categorical,cardinality_threshold)
categorical_columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import time

def pro_bivariate_analysis(df, numerical_cols, categorical_cols, sleep_time=0.5):
    total_plots = len(numerical_cols) * len(categorical_cols)
    plot_rank = 1
    print(f'🔎 Total Bivariate Hist/Box Plots To Generate: {total_plots}')
    
    for cat in categorical_cols:
        if df[cat].nunique() <= 1:
            print(f"Skipping {cat} (only 1 unique value)")
            continue
        
        for num in numerical_cols:
            # Skip IDs if needed
            if num in ['ORDERNUMBER', 'ORDERLINENUMBER']:
                continue
                
            print(f'📊 Plot {plot_rank}: {num} vs {cat}')
            try:
                plt.figure(figsize=(12,5))

                # 1) Histplot with Hue
                plt.subplot(1,2,1)
                sns.histplot(df, x=num, hue=cat, kde=True, element="step", palette='Set2')
                plt.title(f'{num} Distribution by {cat}')
                plt.xticks(rotation=45)

                # 2) Boxplot to check outliers per category
                plt.subplot(1,2,2)
                sns.boxplot(data=df, x=cat, y=num, palette='Set3')
                plt.title(f'{num} by {cat}')
                plt.xticks(rotation=45)

                plt.tight_layout()
                plt.show()
                time.sleep(sleep_time)
                plt.close()
                plot_rank += 1

            except Exception as e:
                print(f"⚠️ Cannot plot {num} vs {cat}: {e}")
                continue
# categorical_columns = ['PRODUCTLINE', 'TERRITORY', 'DEALSIZE']
# numerical = ['ORDERNUMBER', 'QUANTITYORDERED', 'PRICEEACH','SALES']
pro_bivariate_analysis(df_sales, ['QUANTITYORDERED', 'PRICEEACH','SALES'], ['PRODUCTLINE', 'TERRITORY', 'DEALSIZE'])
```

```python
def plot_crosstab_heatmap(df, cat1, cat2):
    ctab = pd.crosstab(df[cat1], df[cat2])
    plt.figure(figsize=(10,7))
    sns.heatmap(ctab, annot=True, fmt='d', cmap='YlGnBu')
    plt.title(f'Crosstab Heatmap: {cat1} vs {cat2}')
    plt.tight_layout()
    plt.show()
plot_crosstab_heatmap(df=df_sales, cat1='TERRITORY', cat2='PRODUCTLINE')
```

# Feature Engineering

```python
# --- Feature Engineering Step 1 ---

import numpy as np
import pandas as pd

df_fe = df_sales.copy()  # Work on a copy for safe practice

# ✅ 1. Create DiscountGiven (MSRP - PRICEEACH)
df_fe['DISCOUNT_GIVEN'] = df_fe['MSRP'] - df_fe['PRICEEACH']

# ✅ 2. Create HighDealFlag (Large Deal = 1, else 0)
df_fe['HIGH_DEAL_FLAG'] = np.where(df_fe['DEALSIZE'] == 'Large', 1, 0)

# ✅ 3. Create isClassicCarDeal (Flag if productline is Classic Cars)
df_fe['IS_CLASSIC_CAR_DEAL'] = np.where(df_fe['PRODUCTLINE'] == 'Classic Cars', 1, 0)

# ✅ 4. Bin QuantityOrdered into Low / Medium / High buckets
df_fe['QUANTITY_BUCKET'] = pd.cut(df_fe['QUANTITYORDERED'],
                                  bins=[0, 30, 60, np.inf],
                                  labels=['Low', 'Medium', 'High'])

# ✅ 5. Bin Sales into Low / Medium / High (you can adjust bins as per your boxplot outlier analysis)
df_fe['SALES_BUCKET'] = pd.cut(df_fe['SALES'],
                               bins=[0, 5000, 10000, np.inf],
                               labels=['Low', 'Medium', 'High'])

# ✅ 6. Interaction Feature: Estimated Sales = PRICEEACH * QUANTITYORDERED
df_fe['ESTIMATED_SALES'] = df_fe['PRICEEACH'] * df_fe['QUANTITYORDERED']

print("✅ Feature Engineering Step 1 complete! New columns added:")
print(df_fe.columns)
```

```python
# --- Feature Engineering Step 2: Encoding ---

from sklearn.preprocessing import LabelEncoder

df_fe_encoded = df_fe.copy()  # Work on a copy

# ✅ Label Encoding for Ordinal Columns
le_dealsize = LabelEncoder()
df_fe_encoded['DEALSIZE_ENCODED'] = le_dealsize.fit_transform(df_fe_encoded['DEALSIZE'])

# ✅ One-Hot Encoding for Nominal Categorical Columns
df_fe_encoded = pd.get_dummies(df_fe_encoded, 
                               columns=['TERRITORY', 'PRODUCTLINE'], 
                               drop_first=True)

print("✅ Feature Engineering Step 2 complete! Encoded columns added:")
print(df_fe_encoded.columns)
```
