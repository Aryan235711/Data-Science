# LOGGING

```python
import logging
```

```python
# Configure the basic logging settings
logging.basicConfig(level=logging.DEBUG)

# Log messages
logging.debug('This is a debug message')
logging.log('This is an info message')
logging.warning('This is a warning message')
logging.error('This is an error message')
logging.critical('This is a critical message')
```

```python
# Saving the message in a log file
logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(levelname)s-%(message)s',
    datefmt='%y-%m-%d %H:%M:%S'
                    )

# Log messages
logging.debug('This is a debug message')
logging.info('This is an info message')
logging.warning('This is a warning message')
logging.error('This is an error message')
logging.critical('This is a critical message')
```

## Create multiple loggers for different parts of application

```python
import logging

# Create a logger for module 1
logger1 = logging.getLogger('module1')
logger1.setLevel(logging.DEBUG)

# Create logger for module 2
logger2 = logging.getLogger('module2')
logger2.setLevel(logging.WARNING)

logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s - %(levelname)s - %(message)s',  # Corrected format
    datefmt='%y-%m-%d %H:%M:%S'
)

logger1.debug('This is a debug message')
logger2.warning('This is a warning message')
```
